```python
#Question 01
def reverse_and_concatenate(A, B):
    reversed_A = A[::-1]
    result = reversed_A + B
    return result

string_A = "Hello"
string_B = " Python!"
final_string = reverse_and_concatenate(string_A, string_B)
print(final_string)
```
